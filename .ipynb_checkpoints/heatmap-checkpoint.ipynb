```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt 
import shapely
from shapely.geometry import Point, Polygon 
import folium
from folium.plugins import BeautifyIcon
```

```python
df_boligsiden = pd.read_csv(r'C:\Users\danie\OneDrive\Documents\GitHub\assignment1\boligsiden_api_moritz\data\boligsiden_10000.csv')
geo = [Point(xy) for xy in zip(df_boligsiden['longitude'], df_boligsiden['latitude'])]
geo_df = gpd.GeoDataFrame(df_boligsiden,geometry=geo)
```

```python
# create background map 

cph_lat = 55.676098
cph_lng = 12.568337

cph_map = folium.Map(
    location=[cph_lat,cph_lng],
    zoom_start=11.5,
    tiles="cartodbpositron"
    )

# Add houses in Copenhagen

cph_houses = df_boligsiden[df_boligsiden['municipalityCode'] == 101] 

for index, row in cph_houses.iterrows():
    lng = row.longitude
    lat = row.latitude  
    folium.CircleMarker(
      [lat,lng],
      radius=1,
      color='#9768D1',
      fill=True,
      fill_color='#9768D1',
      fill_opacity=1
    ).add_to(cph_map)

# Add houses in Fredriksberg 

fre_houses = df_boligsiden[df_boligsiden['municipalityCode'] == 147]
    
for index, row in fre_houses.iterrows():
    lng = row.longitude
    lat = row.latitude  
    folium.CircleMarker(
      [lat,lng],
      radius=1,
      color='#4b9963',
      fill=True,
      fill_color='#4b9963',
      fill_opacity=1
    ).add_to(cph_map)
    
    
cph_map   
    
    
```
