##### calculating distance to nearest metro with geopandas 

```python
df_hp = df_hp.reset_index(drop=True)
df_hp['soldDate'] = pd.to_datetime(df_hp['soldDate'])
```

```python
#not the fastest but gets the job done for n \approx 67k 
#note: geopandas distance function only work for projected coordinates
def calc_dist(house_geom):
    """
    input -  a point geometry of a given house
    output-  tuple of the form : (distance to closest metro, address of the given metro stations)
    """
    list_values = list(map(house_geom.distance, metro_stations.geometry))     #distance function from geopandas
    idx_min = np.array(list_values)\
                .argmin()
    return (list_values[idx_min], metro_stations.loc[idx_min].Address)
```

```python
closest_metro_series = df_hp.geometry.apply(calc_dist)    
```

```python
df_hp['closest_metro'] = [x[1] for x in closest_metro_series]
df_hp['dist_closest_metro'] = [x[0] for x in closest_metro_series]
```

```python
import statsmodels.formula.api as smf
```

```python
for col in ['dist_closest_metro', 'rooms', 'size']:
    df_hp_fin[col] = df_hp_fin[col].astype(float)
```

```python
df_hp_fin['year_sold'] = df_hp_fin['soldDate'].dt.year
```

```python
len(df_hp_fin.city.unique())
```

```python
#HC0 er White's originale 
results = smf.ols(formula = "Real_price ~ rooms + size + dist_closest_metro + dist_closest_metro + C(city) + C(year_sold)",
                  data =df_hp_fin )\
             .fit(cov_type = "HC0")\
             .summary()
results
```
