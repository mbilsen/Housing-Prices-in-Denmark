```python
from selenium import webdriver
import requests
import bs4 as bs
import time
```

```python
pages_n = 3
listings_n = pages_n * 20

url_searches_list = []
for i in range(0,pages_n,1):
    url_searches_list.append(f"https://www.wg-gesucht.de/wg-zimmer-in-Koeln.73.0.1.{i}.html?pagination=1&pu=")


listings_url = []
for url in url_searches_list:
    response = requests.get(url, "lxml")
    soup = bs.BeautifulSoup(response.content)
    list_item_soup = soup.find_all("div", class_ = "wgg_card offer_list_item")
    for i in range(0,len(list_item_soup),1):
        temp_link = list_item_soup[i].find("a")["href"]
        listings_url.append(temp_link)
    time.sleep(5)
    
listings_url_final = ["https://www.wg-gesucht.de" + url for url in listings_url]
```
